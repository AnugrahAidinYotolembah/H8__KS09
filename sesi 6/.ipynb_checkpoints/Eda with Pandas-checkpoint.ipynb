# Unexpected Missing Values


```python
'''
sejauh ini kita telah melihat missing values, dan 
non_standard missing values. bagaimana jika kita memiliki tipe
yang tidak terduga ? misalnya jika fitur kita harapkan berubah string
tetapi malah jadi numerik, maka secara teknis ini juga merupakan
missing value

mari kita liat kolom owner occupied
'''
df[['OWN_OCCUPIED']]
```

```python
df[['OWN_OCCUPIED']].isnull()
```

```python
'''
di baris keempat, ada angka 12. respons untuk owner occupied jelas harus
berupa string (Y atau N), jadi tipe numerik ini berupa missing value

cara mengatasi masalah di atas :
- loop pada kolom own_occupied
- kita coba untuk semua entry menjadi integer
- jika tidak dapat menjadi integer, maka kita tau kalau entry
  adalah string, keep going
  
contoh :
'''
cnt = 0
for row in df['OWN_OCCUPIED']:
    try:
        int(row)
        df.loc[cnt, 'OWN_OCCUPIED'] = np.nan
    except ValueError:
        pass
    cnt+=1
```

```python
df.head(9)
```

```python
'''
pada intinya kode looping di atas dan menggunakan np.nan Numpy
itu agar merubah missing value yang darinya entri menjadi integer

'''
```

# Summarizi Missing Values

```python
'''
setelah kita membersihkan missing values, kita mungkin ingin 
melihat summarynya, misalnya kita mungkin ingin melihat jumlah
total missing values setiap feature

contoh :
'''
df.isnull() # untuk mengkonfirmasi mencari missing value dan "na"
```

```python
df.isnull().sum() #untuk melihat jumlah total missing values 
                  #untuk setiap feature bisa tambahkan .sum()
```

```python
'''
di lain waktu kita mungkin ingin melakukan pemeriksaan cepat 
untuk melihat apakah kita memiliki nilai yang hilang sama sekali
'''

df.isnull().values.any()
```

```python
# kita mungkin juga ingin mendapatkan jumlah total missing values

df.isnull().sum().sum()
```

```python
'''
REPLACING
seringkali kalian harus memikirkan bagaimana kalian menangani
missing values

terkadang kalian hanya ingin menghapus baris tersebut, di lain
waktu kalian menggantinya
'''
df['ST_NUM'].fillna(125, inplace=True)
```

```python
'''
kemungkinan besar, kalian mungkin ingin melakukan imputasi 
berbasis lokasi / location based imputation. ini cara kalian
melakukannya. 
'''
df.loc[2, 'ST_NUM'] = 125
```

```python
# cara yang sangat umum untuk mengganti missing values 
# menggunakan median :

median = df['NUM_BEDROOMS'].median()
df['NUM_BEDROOMS'].fillna(median, inplace = True)

df.head(9)
```

## Basic Time Series Manipulation

```python
# #  pertama import libraries yang kita butuhkan dan gunakan 
# untuk membuat date range

from datetime import datetime
```

```python
date_rng = pd.date_range(start = '1/01/2020', end = '1/08/2020', freq = 'H')

'''
date range ini memiliki timestamps dengan frekuensi per jam. 
jika kita memanggil date_rng kita akan melihat seperti berikut :

'''

date_rng
```

```python
# membuat contoh data frame dengan timestamp dan lihat 15 elemen pertama :

df = pd.DataFrame(date_rng, columns=['date'])

df['data'] = np.random.randint(0, 100, size = (len(date_rng)))

df.head()
```

```python
# konversikan indeks data frame menjadi datetime index tampilan
# elemen pertama :

df['datetime'] = pd.to_datetime(df['date'])

df = df.set_index('datetime')

df.drop(['date'], axis =1, inplace= True)

df.head()
```

```python
# ubah date_rng menjadi list of string dan kemudian ubah string
# tersebut menjadi stempel waktu

string_date_rng = [str(x) for x in date_rng]
string_date_rng
```

```python
# mengkonversi string menjadi timestamps dengan melihat formatnya

timestamp_date_rng = pd.to_datetime(string_date_rng,
                                   infer_datetime_format = True)

timestamp_date_rng
```

```python
# membuat daftar berubah" menjadi string dan mengubahnya menjadi
# timestamps

string_date_rng_2 = ['June-01-2020', 'June-02-2020', 'June-03-2020']

timestamp_date_rng_2 = [datetime.strptime(x, '%B-%d-%Y') for x in string_date_rng_2]

timestamp_date_rng_2
```

```python
# masukkan menjadi dataframe

df2 = pd.DataFrame(timestamp_date_rng_2, columns=['date'])

df2
```

```python
# melihat dataframe aslinya

df[df.index.day == 2]
```

```python
# bisa melihat tanggal yang ingin kita lihat index dari dataframe:

df['2020-01-03']
```

```python
df['2020-01-04' : '2020-01-06']
```

```python
# menghitung rata rata harian dari data :

df.resample('D').mean()
```

```python
'''
mari buat kolom baru di df asli kita yang menghitung rolling sum
selama periode 3 window dan kemudian lihat dibagian atas data frame
'''

df['rolling_sum'] = df.rolling(3).sum()
df.head()
```

```python
df['rolling_sum_backfilled'] = df['rolling_sum'].fillna(method = 'backfill')

df.head()
```
