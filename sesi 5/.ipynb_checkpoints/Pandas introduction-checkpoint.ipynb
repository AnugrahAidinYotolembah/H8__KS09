## Understanding series object

### struktur data python yang paling dasar adalah list, yang juga merukapan titik awal yang baik untuk mengenal objek pandas.Series. buat objek series baru menggunakan 



```python
'''
fungsi objek series :

- urutan nilai

- urutan pengidentifikasi yang merupakan indeks
'''

Revenues = pd.Series([555, 7000, 1980])
print(Revenues)
```

```python
# menggunakan .series

print(Revenues.values)

# fungsi dari revenues.values mengembakalikan nilai dalam seri 
```

```python
# menggunakan .index

print(Revenues.index)

# fungsi dari reveues.index mengembalikan indexs posisi
```

```python
'''
fungsi dari revenues.values mengembakalikan nilai dalam seri 

sedangkan

fungsi dari reveues.index mengembalikan indexs posisi
'''
```

```python
'''
meskipun pandas dibuat di atas numpy, perbedaan yang signifikan 
terletak pada pengindeksannya. sama seperti array numpy, pandas
series integer yang ditentukan secara implisit

indeks implisit ini menunjukkan posisi elemen dalam serius

namun serius juga dapat memiliki jenis indeks yang berubah - ubah.
kalian dapt menganggap indeks eksplisit ini sebagai label untuk
baris tertentu

'''

# contoh :

city_revenues = pd.Series(
    [4200, 8000, 64000],
    index = ["Amsterdam", "Toronto", "Tokyo"]
)

print(city_revenues)
```

```python
'''
di sini, indekx adalah daftar nama kota yang diwakili oleh string.
kalian mungkin telah memperhatikan bahwa dictionary python menggunakan
string juga. dan ini adalah analogi yang berguna untuk di ingat. 
kalian dapat menggunakan blok kode di atas untuk membedak antar dua
jenis series : 

- revenues :
  seri ini berperilaku seperti list python karena hanya memiliki 
  indeks posisi
  

- city_revenues :
  seri ini bertindak seperti dictonary python karena menampilkan 
  indeks posisi dan label

'''
```

```python
# buat series dengan indeks label dari dictionary python :

city_employee_count = pd.Series({"palu" : 5, "tokyo" : 7})
print(city_employee_count)
```

```python
# nilai siswa pakai dictionary python :

nilai_siswa = pd.Series({"didi" : 100, "syarun" : 90, "baso" : 80})
print(nilai_siswa)
```

```python
nilai_siswa.keys()


```

```python
"didi" in nilai_siswa
```

```python
city_data = pd.DataFrame({
    "revenue" : city_revenues,
    "employee_count" : city_employee_count
})

print(city_data)
```

```python
'''daftar nama kelas C'''

kelas_c = pd.Series({
    "didi" : "kelas C",
    "syarun" : "kelas C",
    "dival" : "kelas C",
})
print(kelas_c)
```

```python
'''daftar nilai kelas C'''

nilai_kelas_c = pd.Series([100, 100, 90])
print(nilai_kelas_c)
```

```python
siswa_data = pd.DataFrame({
    "data siswa" : kelas_c,
    "nilai kelas C" : nilai_kelas_c
})

print(siswa_data)
```

```python
# percobaan 2

# nama handphone,harga handphone, tipe handphone

# nama handphone
nama_handphone = pd.Series(
    ['oppo', 'samsung', 'iphone']
    

)
print(nama_handphone)
```

```python
# harga handphone

harga_handphone = pd.Series([10000, 20000, 13000])

print(harga_handphone)
```

```python
# tipe handphone

tipe_handphone = pd.Series(['android', 'android', 'ios'])

print(tipe_handphone)
```

```python
# untuk membuat dataframe

jual_handphone = pd.DataFrame({
    "nama handphone" : nama_handphone,
    "harga handphone" : harga_handphone,
    "tipe_handphone" : tipe_handphone
})

print(jual_handphone)
```

```python
# keys untuk mengethui indeks yang berada di harga_handphone

harga_handphone.keys
```
