## Understanding series object

### struktur data python yang paling dasar adalah list, yang juga merukapan titik awal yang baik untuk mengenal objek pandas.Series. buat objek series baru menggunakan list



```python
'''
fungsi objek series :

- urutan nilai

- urutan pengidentifikasi yang merupakan indeks
'''

Revenues = pd.Series([555, 7000, 1980])
print(Revenues)
```

```python
# menggunakan .series

print(Revenues.values)

# fungsi dari revenues.values mengembakalikan nilai dalam seri 
```

```python
# menggunakan .index

print(Revenues.index)

# fungsi dari reveues.index mengembalikan indexs posisi
```

```python
'''
fungsi dari revenues.values mengembakalikan nilai dalam seri 

sedangkan

fungsi dari reveues.index mengembalikan indexs posisi
'''
```

```python
'''
meskipun pandas dibuat di atas numpy, perbedaan yang signifikan 
terletak pada pengindeksannya. sama seperti array numpy, pandas
series integer yang ditentukan secara implisit

indeks implisit ini menunjukkan posisi elemen dalam serius

namun serius juga dapat memiliki jenis indeks yang berubah - ubah.
kalian dapt menganggap indeks eksplisit ini sebagai label untuk
baris tertentu

'''

# contoh :

city_revenues = pd.Series(
    [4200, 8000, 64000],
    index = ["Amsterdam", "Toronto", "Tokyo"]
)

print(city_revenues)
```

```python
# menggunakan .loc dan .iloc

'''
operator pengindeksian ([]) memang mudah digunakan, tetapi ada
peringatan. bagaimana juka labelnya juga angka ? katakanlah 
kalian harus bekerja dengan objek series seperti ini
'''

colors = pd.Series(
    ["red", "purpel", "blue", "green", "yello"],
    index = [1, 2, 3, 4, 8]
)

print(colors)

# .loc mengacu pada indeks label

# .iloc mengacu pada indeks posisi
```

```python
'''
 .loc mengacu pada indeks label
  kalau indeks label hitungannya tidak di mulai dengan angka 0

 .iloc mengacu pada indeks posisi
  kalua indeks posisi dimulai dengan angka 0 
'''


colors = pd.Series(
    ["red", "purple", "blue", "green", "yellow"],
    index = [1, 2, 3, 4, 8]
)

print(colors)
```

```python
colors.loc[1]

# colors.loc[1] mengembalikan "merah"
# colors.iloc[1] mengembalikan "ungu", elemen dengan indeks 1
```

```python
colors.iloc[1]
```

```python
'''
.loc dan .iloc juga mendukung fitur yang kalian harapkan dari 
operator pengindeksan, seperti slicing. namun metode akses data
ini memiliki perbedaan penting.  

.iloc mengecualikan elemen penutup

.loc menyertakan elemen penutup


'''

colors.iloc[1:3]
```

```python
colors.loc[3:8]
```

```python
# kalian juga dapat menyertakan indeks posisi negatif ke .iloc

colors.iloc[-2]
```

```python
'''
you can use .iloc on a series similiar to using [] on a list

you can use .loc on a series similiar to using [] on a dictionary
'''
```
