# Area Plots

```python
'''
di bagian terakhir, kita membuat plot garis yang menvisualisikan 
5 negara teratas yang menyumbang paling banyak imigran ke kanada
dari 1980 hingga 2013. dengan sedikit modifikasi pada kode,
kita dapat menvisualisasikan plot ini sebagai cumulative plot
juga di kenal sebagai stacked line plot atau area plot
'''
df_can.sort_values(['Total'], ascending =False, axis = 0, inplace = True)

#get the top 5 entries
df_top5 = df_can.head()

#transpose the dataframe
df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
'''
Area plots ter-stacked secara default. dan untuk menghasilkan 
stacked area plot, setiap kolom harus berupa semua nilai positif
atau semua nilai negatif (nilai NaN akan ditetapkan secara default ke 0)
untuk menhasilkan unstacked plot, berikan stacked = False
'''
```

```python
#let's change the index values of df_top5 to type integer for plotting
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind = 'area', stacked = False, figsize = (20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
'''
plot unstacked memiliki transparansi default(alpha value)
sebesar 0,5. kita dapat mengubah nilai ini dengan memasukkan
parameter alpha.
'''
```

```python


df_top5.plot(kind = 'area', alpha = 0.25, stacked = False, figsize = (20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
'''
TWO TYPES OF PLOTTING

ada dua gaya / pilihan membuat plot dengan matplotlib. membuat
plot menggunakan artist layer dan membuat plot menggunakan
scripting layer

kita dapat menggunakan plt yaitu matplotlib.pyplot dan menambahkan
lebih banyak elemen dengan memanngil metode yang berbeda secara 
prosedural misalnya plt.title(...) untuk menambahkan judul
atau plt.xlabel(...) untuk menambahkan label sumbu x


'''

```

```python
# 1. tipe plotting pada umumnya

df_top5.plot(kind = 'area', alpha = 0.35, figsize = (20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
# 2. tipe mengunakan format ax.set



ax = df_top5.plot(kind = 'area', alpha = 0.25, figsize = (20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
'''
Question: Use the scripting layer to create a stacked area 
plot of the 5 countries that contributed the least to immigration 
to Canada from 1980 to 2013. Use a transparency value of 0.45.

'''

df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()

df_least5.plot(kind = 'area', alpha = 0.45, stacked = True, figsize = (20, 10),)

plt.title('Immigration trend of least 5 countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
ax = df_least5.plot(kind = 'area', alpha = 0.35, figsize = (20, 10),)

ax.set_title('Immigration trend of least 5 countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

# HISTOGRAM 

```python
'''
Histogrram adalah cara untuk mempresentasikan distribusi frekuensi
dari dataset numerik. cara kerjanya adalah mempertisi sumbu x menjadi
beberapa bin. menetapkan setiap titik data dalam ke setiap bin.
jadi sumbu y adalah frekuensi atau banyaknya titik data pada setiap bin
note

Question: What is the frequency distribution of the number 
          (population) of new immigrants from the various 
          countries to Canada in 2013?


'''

df_can['2013'].head()
```

```python
# np.histrogram returns 2 values

count, bin_edges = np.histogram(df_can['2013'])
print(count)
print(bin_edges)
```

```python
'''
secara default, metode histogram memecah dataset menjadi 10 bin
gambar dibawah ini merangkum kisaran bin dan distribusi frekuensi
imigrasi di tahun 2013. dapat kita lihat tahun 2013 :

- 178 countries contributed between 0 to 3412.9 imigrants
- 11 countries contributed between 3412.9 to 6825.8 immigrants
- 1 country contributed between 6285.8 to 10238.7 immigraants and so


kita dapat dengn mudah membuat grafik distribusi ini dengan 
meneruskan kind = hist ke plot ()
'''

df_can['2013'].plot(kind='hist', figsize=(8, 5))

plt.title('Histrogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
count, bin_edges = np.histogram(df_can['2013'])

df_can['2013'].plot(kind = 'hist', figsize = (8, 5), xticks = bin_edges)

plt.title('Histrogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

# Buble plots

```python
df_can_t = df_can[years].transpose() # transposed dataframe

# cast the Years (the index) to type int
df_can_t.index = map(int, df_can_t.index)

# let's label the index. This will automatically be the column name when we reset the index
df_can_t.index.name = 'Year'

# reset index to bring the Year in as a column
df_can_t.reset_index(inplace=True)

# view the changes
df_can_t.head()
```

```python
# normalize Brazil data
norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) / (df_can_t['Brazil'].max() - df_can_t['Brazil'].min())

# normalize Argentina data
norm_argentina = (df_can_t['Argentina'] - df_can_t['Argentina'].min()) / (df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

```python
# Brazil
ax0 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,                  # transparency
                    color='green',
                    s=norm_brazil * 2000 + 10,  # pass in weights 
                    xlim=(1975, 2015)
                   )

# Argentina
ax1 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 2000 + 10,
                    ax = ax0
                   )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

```python
norm_china = (df_can_t['China'] - df_can_t['China'].min()) / (df_can_t['China'].max() - df_can_t['China'].min())
norm_india = (df_can_t['India'] - df_can_t['India'].min()) / (df_can_t['India'].max() - df_can_t['India'].min())
```

```python
ax0 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='China',
                    figsize=(14, 8),
                    alpha=0.5,                  # transparency
                    color='green',
                    s=norm_china * 1300 + 10,  # pass in weights 
                    xlim=(1975, 2015)
                   )

ax1 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='India',
                    alpha=0.5,
                    color="blue",
                    s=norm_india * 1300 + 10,
                    ax = ax0
                   )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from China and India from 1980 - 2013')
ax0.legend(['China', 'India'], loc='upper left', fontsize='x-large')
```
